```python
import pandas as pd
import matplotlib as plt
```

Data Preparation

```python
## Load the data
crypto = "crypto_data.csv"
crypto_data = pd.read_csv(crypto, index_col=0)
crypto_data
```

```python
## Discard all cryptocurrencies that are not being traded

crypto_df = crypto_data[crypto_data['IsTrading'] == True]
crypto_df
```

```python
## Finding null values
crypto_df.isnull().sum()
```

```python
## Drop the null rows from the dataframe
crypto_df.dropna()
```

```python
## Filter for cryptocurrencies that have been mined. That is, the total coins mined should be greater than zero

mined_crypto = crypto_df[crypto_df["TotalCoinsMined"]>0]
mined_crypto
```

```python
## Drop the column CoinName
mined_crypto = mined_crypto.drop(['CoinName'], axis=1)
mined_crypto
```

```python
## create dummy variables
X = pd.get_dummies(data=mined_crypto, columns=["Algorithm", "ProofType"])
X.head()
```

```python
## Drop the column IsTrading
X = X.drop(['IsTrading'], axis=1)
X
```

```python
## Standardize your dataset
from sklearn.preprocessing import StandardScaler, MinMaxScaler
X = StandardScaler().fit_transform(X)
print(X[0:5])
```

Dimensionality Reduction

```python
# Initialize PCA model
from sklearn.decomposition import PCA
pca = PCA(n_components=.9)

# Get two principal components for the iris data.
X_pca = pca.fit_transform(X)
print(X_pca.shape)
```

```python
# Transform PCA data to a DataFrame
df_X_pca = pd.DataFrame(
    data=X_pca, columns=["prin component 1", "prin component 2","prin component 3"]
)
df_X_pca.head()
```

```python
# Fetch the explained variance
pca.explained_variance_ratio_
```

```python
# Initialize t-SNE model
from sklearn.manifold import TSNE
tsne = TSNE(learning_rate=35)
```

```python
# Reduce dimensions
tsne_features = tsne.fit_transform(X_pca)
```

```python
# The dataset has 2 columns
tsne_features.shape
```

```python
crypto_tsne = pd.DataFrame(data=tsne_features,columns=["test","testX"])
```

```python
# Prepare to plot the dataset

# The first column of transformed features
crypto_tsne['x'] = tsne_features[:,0]

# The second column of transformed features
crypto_tsne['y'] = tsne_features[:,1]
```

```python
# Visualize the clusters
import matplotlib.pyplot as plt
plt.scatter(crypto_tsne['x'], crypto_tsne['y'])
plt.show()
```

```python
from sklearn.cluster import KMeans
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range ok k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(crypto_tsne)
    inertia.append(km.inertia_)

# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```
