```python
import pandas as pd
import matplotlib as plt
```

Data Preparation

```python
crypto = "crypto_data.csv"
crypto_data = pd.read_csv(crypto, index_col=0)
crypto_data
```

```python
## Discard all cryptocurrencies that are not being traded

crypto_df = crypto_data[crypto_data['IsTrading'] == True]
crypto_df
```

```python
## Finding null values
crypto_df.isnull().sum()
```

```python
## Drop the null rows from the dataframe
crypto_df.dropna()
```

```python
## Filter for cryptocurrencies that have been mined. That is, the total coins mined should be greater than zero

mined_crypto = crypto_df["TotalCoinsMined"]>0
mined_crypto
```
